```python
%load_ext autoreload
%autoreload 2
import generator
import time
import sys 
from arcworld.general_utils import plot_task, plot_grid
```

```python
# At each run, the generator will create a new task with a different combination of transformations, as described in the config file.

config = {
    "min_n_shapes_per_grid": 1,
    "max_n_shapes_per_grid": 1, 
    "n_examples": 1,
    "min_grid_size": 16,
    "max_grid_size": 16,
    "allowed_combinations": None,
    "allowed_transformations": ["translate_up"],
    "min_transformation_depth": 1,
    "max_transformation_depth": 1,
    "shape_compulsory_conditionals": ["is_shape_less_than_8_rows", 
                                      "is_shape_less_than_8_cols",
                                      "is_shape_fully_connected",
                                      "is_shape_not_hollow"
                                      ],
}

gen = generator.generator(config, debug_mode=True) # Initialize the generator with the config
task = gen.generate_single_task() # Generate a single task
plot_task(task)
```

```python
import os
import numpy as np
from tqdm import tqdm

saving_path = "./objects_benni/"
os.makedirs(saving_path, exist_ok=True)

for i in range(900000):
    task = gen.generate_single_task()
    input_grid = task["pairs"][0]["input"]  # numpy array
    file_path = os.path.join(saving_path, f"input_{i+100000}.npy")
    np.save(file_path, input_grid)
    if i%1000 == 0:
        print("generated ", i)
```
